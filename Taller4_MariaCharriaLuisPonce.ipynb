##Metropolis Hasting

```python
#1
#1a
n = 10  
r = 7  
def prior(p):
    if 0 < p < 1:
        return 1
    else:
        return 0

def binomial(p):
    return p ** r * (1 - p) ** (n - r)

def metropolis(n_iterations, initial_value, proposal_sd):
    current = initial_value
    samples = [current]

    for _ in range(n_iterations):
        proposal = np.random.normal(current, proposal_sd)
        acceptance_ratio = min(1, (binomial(proposal) * prior(proposal)) / (binomial(current) * prior(current)))
        if np.random.uniform(0, 1) < acceptance_ratio:
            current = proposal
        samples.append(current)
    return samples

n_iterations = 10000  # Número de iteraciones del algoritmo
initial_value = 0.5  # Valor inicial para pˆ
proposal_sd = 0.1  # Desviación estándar de la propuesta

samples = metropolis(n_iterations, initial_value, proposal_sd)
max_posterior = max(samples)
print("a) El parámetro asociado a la probabilidad de éxito p^ es:", round(max_posterior,5))

#1b
binomial_variance = (r * (n - r)) / (n ** 2 * (n - 1))
lower_quantile = np.percentile(samples, (100 - 68) / 2)
upper_quantile = np.percentile(samples, 100 - (100 - 68) / 2)
lower_error = max_posterior - lower_quantile
upper_error = upper_quantile - max_posterior
print("b) Errores asociados al parámetro p^ a un nivel de confianza del 68%:", round(lower_error,5), round(upper_error,5))

#1c
print("c) ¿Podemos decir que la moneda está truncada?")
if 0 < max_posterior < 1:
    print("No hay suficiente evidencia para afirmar que la moneda está truncada.")
else:
    print("Existe evidencia de que la moneda está truncada.")
```

```python
#3

mu = 2 
sigma = 0.5 

def normal_pdf(x):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2) / (sigma * np.sqrt(2 * np.pi))

def metropolis_sampling(n_iterations, initial_value, proposal_sd):
    current = initial_value
    samples = [current]

    for _ in range(n_iterations):
        proposal = np.random.normal(current, proposal_sd)

        acceptance_ratio = min(1, normal_pdf(proposal) / normal_pdf(current))

        if np.random.uniform(0, 1) < acceptance_ratio:
            current = proposal

        samples.append(current)

    return samples

N = 1000  # Número de eventos a muestrear
initial_value = 0  # Valor inicial
proposal_sd = 1  # Desviación estándar de la propuesta

samples = metropolis_sampling(N, initial_value, proposal_sd)

plt.hist(samples, bins=30, density=True, alpha=0.5)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x, normal_pdf(x), 'r-', lw=2)
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.title('Muestreo de una distribución normal utilizando el algoritmo de Metrópolis')
plt.show()
```

```python
#4

def cauchy_pdf(x):
    return 1 / (np.pi * (1 + x ** 2))

def metropolis_sampling(n_iterations, initial_value, proposal_sd):
    current = initial_value
    samples = [current]

    for _ in range(n_iterations):
        proposal = np.random.normal(current, proposal_sd)

        acceptance_ratio = min(1, cauchy_pdf(proposal) / cauchy_pdf(current))

        if np.random.uniform(0, 1) < acceptance_ratio:
            current = proposal

        samples.append(current)

    return samples

N = 1000  # Número de eventos a muestrear
initial_value = 0  # Valor inicial
proposal_sd = 1  # Desviación estándar de la propuesta

samples = metropolis_sampling(N, initial_value, proposal_sd)

plt.hist(samples, bins=30, density=True, alpha=0.5)
x = np.linspace(-10, 10, 100)
plt.plot(x, cauchy_pdf(x), 'r-', lw=2)
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.title('Muestreo de una distribución estándar de Cauchy utilizando el algoritmo de Metrópolis')
plt.show()
```

## Parameter estimation

```python
#3

def prior_uniform(mu, sigma):
    if 3 <= mu <= 5 and 0.5 <= sigma <= 3.5:
        return 1
    else:
        return 0

def gaussian_likelihood(x, mu, sigma):
    exponent = -((x - mu) ** 2) / (2 * sigma ** 2)
    likelihood = (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(exponent)
    return likelihood

def calculate_log_posterior(x, mu, sigma):
    prior = prior_uniform(mu, sigma)
    likelihood = gaussian_likelihood(x, mu, sigma)
    log_posterior = np.log(prior) + np.log(likelihood)
    return log_posterior

def metropolis_algorithm(x, n_samples, initial_mu, initial_sigma, step_size):
    samples = []
    mu_current = initial_mu
    sigma_current = initial_sigma
    log_posterior_current = calculate_log_posterior(x, mu_current, sigma_current)
    
    for _ in range(n_samples):
        mu_proposed = np.random.normal(mu_current, step_size)
        sigma_proposed = np.random.normal(sigma_current, step_size)
        log_posterior_proposed = calculate_log_posterior(x, mu_proposed, sigma_proposed)
        
        acceptance_ratio = np.exp(log_posterior_proposed - log_posterior_current)
        
        if np.all(np.random.uniform() < acceptance_ratio):
          mu_current = mu_proposed
          sigma_current = sigma_proposed
          log_posterior_current = log_posterior_proposed
        
        samples.append((mu_current, sigma_current))
    
    return samples

url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/Gaussiano.csv'
data = np.loadtxt(url, dtype=float, skiprows=1)

x = data

n_samples = 2 * 10**4
initial_mu = 4.0
initial_sigma = 2.5
step_size = 0.1

samples = metropolis_algorithm(x, n_samples, initial_mu, initial_sigma, step_size)

mu_samples = np.array([sample[0] for sample in samples])
sigma_samples = np.array([sample[1] for sample in samples])

mu_estimated = np.mean(mu_samples)
sigma_estimated = np.mean(sigma_samples)

print("Mejor valor estimado de mu (μ^):", mu_estimated)
print("Mejor valor estimado de sigma (σ^):", sigma_estimated)


plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.hist(mu_samples, color="dodgerblue", bins=50, density=True, alpha=0.5)
plt.axvline(x=mu_estimated, color='r', linestyle='--', label='Estimated mu')
plt.xlabel('mu')
plt.ylabel('Density')
plt.legend()
plt.subplot(2, 1, 2)
plt.hist(sigma_samples,color="dodgerblue", bins=50, density=True, alpha=0.5)
plt.axvline(x=sigma_estimated, color='r', linestyle='--', label='Estimated sigma')

sorted_sigma_samples = np.sort(sigma_samples)

lower_index = int(n_samples * 0.16)
upper_index = int(n_samples * 0.84)

sigma_lower = sorted_sigma_samples[lower_index]
sigma_upper = sorted_sigma_samples[upper_index]

sigma_minus = sigma_estimated - sigma_lower
sigma_plus = sigma_upper - sigma_estimated

print("Error sigma-:", sigma_minus)
print("Error sigma+:", sigma_plus)
```
